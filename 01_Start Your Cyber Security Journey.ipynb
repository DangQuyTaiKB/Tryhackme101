## Task 2: Areas of Defensive Security
In this task, we are being briefed regarding two main topics related to defensive security and one of their respective tasks:
- Security Operations Center (SOC), covering Threat Intelligence
- Digital Forensics and Incident Response (DFIR), covering Malware Analysis

### Security Operations Center (SOC)
A Security Operations Center (SOC) is a team of cyber security professionals that monitors the network and its systems to detect malicious cyber security events. 
Key areas of focus:
- Vulnerabilities: Identifying and mitigating system weaknesses
- Policy Violations: Monitoring for non-compliance with security rules
- Unauthorized Activity: Detecting and blocking unauthorized access attempts
- Network Intrusions: Identifying and responding to network breaches
- Threat Intelligence: Gathering and analyzing information about potential threats

### Threat Intelligence
- Information about potential threats to a system or organization
- Helps organizations prepare for and respond to cyberattacks
- Involves collecting, processing, and analyzing data on threats and adversaries
- Identifies potential threats and their tactics, techniques, and procedures
- Enables organizations to implement proactive security measures and respond effectively to attacks

### Digital Forensics and Incident Response (DFIR)

#### Digital Forensics
- Application of scientific methods to investigate digital devices
- Focuses on analyzing digital evidence to identify and prosecute cybercrimes
- Key areas of investigation:
    - File systems: Analyzing file systems for deleted, modified, or hidden data
    - System memory: Examining memory dumps for active processes and malicious code
    - System logs: Analyzing log files for system events, user activity, and security alerts
    - Network traffic: Examining network traffic for suspicious activity and data exfiltration

#### Incident Response
- A systematic approach to handling security incidents
- Involves identifying, analyzing, containing, eradicating, and recovering from security breaches
- Key phases:
    - Preparation: Developing incident response plans, training teams, and establishing procedures
    - Detection and Analysis: Identifying and analyzing security incidents
    - Containment, Eradication, and Recovery: Isolating infected systems, removing malware, and restoring systems to normal operation
    - Post-Incident Activity: Documenting the incident, conducting lessons learned, and implementing corrective measures

#### Malware Analysis
- The process of examining malicious software to understand its behavior and functionality
- Techniques:
    - Static analysis: Examining malware code without executing it
    - Dynamic analysis: Executing malware in a controlled environment to observe its behavior
- Types of Malware
    - Viruses: Self-replicating code that spreads through systems
    - Worms: Self-propagating malware that spreads without human interaction
    - Trojans: Malicious programs disguised as legitimate software
    - Ransomware: Malware that encrypts data and demands a ransom for decryption
    - Spyware: Malware that collects sensitive information without user consent

## Task 2: Evaluating Information on the Internet

The Internet is a vast resource where anyone can publish their thoughts, be it through blogs, articles, or social media posts. This open platform allows a mix of both credible information and unfounded claims. When evaluating information, consider the following:

- **Source**: Identify the author or organization and their credibility on the subject.
- **Evidence and Reasoning**: Look for credible evidence supporting the claims.
- **Objectivity and Bias**: Check for impartiality and multiple perspectives.
- **Corroboration and Consistency**: Verify the information by comparing it with reliable, independent sources.

## Task 3: Search Engines

We all use search engines, but advanced techniques can greatly refine results. Google, Bing, and DuckDuckGo offer several operators:

- **"exact phrase"**: Finds pages with the exact phrase, e.g., `"passive reconnaissance"`.
- **`site:`**: Limits the search to a specific domain, e.g., `site:tryhackme.com success stories`.
- **`-`**: Excludes terms, e.g., `pyramids -tourism`.
- **`filetype:`**: Searches for specific file types, e.g., `filetype:pdf cyber warfare report`.

For more advanced search operators, check the help documentation of your preferred search engine.

# Task 5: Vulnerabilities and Exploits

### CVE
The Common Vulnerabilities and Exposures (CVE) program serves as a dictionary of vulnerabilities, providing standardized identifiers for security issues in software and hardware products. Each vulnerability is assigned a CVE ID in a format like `CVE-2024-29988`. This unique ID ensures that everyone, from security researchers to vendors, refers to the same vulnerability consistently. For instance, CVE-2024-29988 would denote a specific security issue.

The CVE system is maintained by the MITRE Corporation. To search for CVEs, you can visit the [CVE Program website](https://cve.mitre.org) or the [National Vulnerability Database (NVD)](https://nvd.nist.gov). Below is an example screenshot of CVE-2014-0160, also known as Heartbleed.

### Exploit Database
Exploit databases are essential for security assessments, especially for red team exercises where you assess a company's defenses by simulating attacks. However, remember to only exploit vulnerabilities when legally authorized.

To locate a working exploit code, the [Exploit Database](https://www.exploit-db.com) is a valuable resource, offering tested and verified exploit codes. Additionally, [GitHub](https://github.com) often hosts proof-of-concept (PoC) and exploit codes related to CVEs, as shown below in a screenshot of search results for Heartbleed.

# Task 6: Technical Documentation

### Linux Manual Pages
Before the internet was widely available, Unix-like systems used manual (man) pages to document commands. Each command on Linux or Unix-like systems has a corresponding man page, covering usage, options, and examples. To view the `ip` command's manual, for example, use `man ip`. You can also search for `man ip` in your preferred search engine to view the documentation online.

The AttackBox in TryHackMe provides an in-browser Linux terminal. You can start the AttackBox and run commands like `man ip`, and exit the manual by pressing `q`.
